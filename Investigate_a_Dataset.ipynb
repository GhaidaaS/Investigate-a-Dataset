### Data Cleaning (dropping any useless data without any affect to other data)

```python
# After discussing the structure of the data and any problems that need to be
#   cleaned, perform those cleaning steps in the second part of this section.
columns=["imdb_id","homepage","keywords"]
df.drop(columns,axis=1,inplace=True)
df.dropna(inplace=True)
df.isnull().sum()
```

```python
## display the describtion of data 
df.describe()
## drop zero value from runtime
df=df.drop(df.index[df['runtime'] == 0].tolist())
## display the describtion of data to make sure thet the cleaning is correct
df.describe()
```

```python
## method to replace zero value with mean
def replace0WithMean (c,s):
    ## find nonzero mean
    nonzero_mean = df[ c != 0 ].mean()
    ## replace the zero values with this mean 
    df.loc[ c == 0, s] = nonzero_mean
```

```python
## replace zero value in the following columns with mean
replace0WithMean(df.revenue,'revenue')
replace0WithMean(df.budget,'budget')
replace0WithMean(df.budget_adj,'budget_adj')
replace0WithMean(df.revenue_adj,'revenue_adj')
## display the describtion of data to make sure thet the cleaning work
df.describe()
```

```python
directorm=df.groupby(df.director.tolist()).size().reset_index().rename(columns={0:"Counter"})
directorm=directorm.sort_values("Counter",ascending=False).head(20)

plt.plot(directorm['Counter'],directorm['index'])
plt.title('The most pupular directors ')
plt.xlabel('director',fontsize=17)
plt.ylabel('Counter',fontsize=17)
```

 From the sampel i choose Woody allen is the most pupular directors

```python
genresn=df.groupby([df.genres.tolist(),df.runtime]).size().reset_index().rename(columns={0:"Counter"})
genresn=genresn.sort_values("Counter",ascending=False).head(250)
genresn

plt.bar(genresn['level_0'],genresn['runtime'])
plt.title('The relation between runtime and generes of movies')
plt.xticks(rotation=90)
plt.xlabel('Genres',fontsize=17)
plt.ylabel('Runtime',fontsize=17)
```

### Research Question 1 (What is the most popular generes)

```python
# Use this, and more code cells, to explore your data. Don't forget to add
#   Markdown cells to document your observations and findings.

genresm=df.groupby(df.genres.tolist()).size().reset_index().rename(columns={0:"Counter"})
genresm=genresm.sort_values("Counter",ascending=False).head(35)
genresm

plt.bar(genresm['index'],genresm['Counter'])
plt.title('Generes and number of it ')
plt.xticks(rotation=90)
plt.xlabel('Genres',fontsize=17)
plt.ylabel('Counter',fontsize=17)
```
